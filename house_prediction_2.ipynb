# Create test set

```python
from sklearn.model_selection import train_test_split
X = data.drop(['Price'], axis=1)
y = data.Price
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=42)
```

```python
# put your label back 
train = pd.concat([X_train, y_train], axis=1)
valid = pd.concat([X_valid, y_valid], axis=1)
```

# Numerical features


```python
#get numeric columns 
#df.select_dtypes(exclude='o')
numeric_columns = [col for col in train.columns if train[col].dtype in ['float64', 'int64']]

num_train = train[numeric_columns]
num_valid = valid[numeric_columns]

print('Numeric columns: {}'.format((numeric_columns)))
```

```python
# columns with high skew
high_skew_cols = [col for col in num_train.columns if num_train[col].skew() > 2]
print('Skewed columns: {}'.format(high_skew_cols))
```

```python
# we wil use quantile method. 
num_train.Price.quantile(0.1)
num_train.Price.quantile(0.9)
```

```python
import numpy as np
# fix outliers
for col in high_skew_cols:
    #train data
    num_train[col] = np.where(num_train[col] < num_train[col].quantile(0.1), num_train[col].quantile(0.1), num_train[col])
    num_train[col] = np.where(num_train[col] > num_train[col].quantile(0.9), num_train[col].quantile(0.9), num_train[col])
    #fix outliers for test data with train
    num_valid[col] = np.where(num_valid[col] < num_train[col].quantile(0.1), num_valid[col].quantile(0.1), num_valid[col])
    num_valid[col] = np.where(num_valid[col] > num_train[col].quantile(0.1), num_valid[col].quantile(0.9), num_valid[col])
num_train.skew().sort_values(ascending=False)
```

```python
# missing values
num_train.isnull().sum()
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')

final_num_train = pd.DataFrame(imputer.fit_transform(num_train))
final_num_valid = pd.DataFrame(imputer.transform(num_valid))

#put columns back
final_num_train.columns = numeric_columns
final_num_valid.columns = numeric_columns
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# Function for comparing different approaches
def score_dataset(X_train, X_valid, y_train, y_valid):
    model = RandomForestRegressor(n_estimators=10, random_state=0)
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid, preds)
```
