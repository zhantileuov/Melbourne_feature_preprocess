# Create test set

```python
from sklearn.model_selection import train_test_split
X = data.drop(['Price'], axis=1)
y = data.Price
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)
```

# Vizualize and discover

```python
melb_house_train = pd.concat([X_train, y_train], axis=1)
melb_house_valid = pd.concat([X_valid, y_valid], axis=1)
melb_house_train.plot(kind='scatter', x = 'Longtitude', y = 'Lattitude', 
                alpha=0.15, c = 'Price', cmap=plt.get_cmap("jet"),
               figsize=(10,7))
plt.show()
```

```python
a_train = num_features_train.copy()
a_valid = num_features_valid.copy()
a_train.skew()
```

```python
print(a_train.Price.quantile(0.1))
print(a_train.Price.quantile(0.9))
```

```python
# Quantile-based Flooring and Capping (replacing outliers)
import numpy as np
for i, j in a_train.skew().items():
    if j > 1.8:
        a_train[i] = np.where(a_train[i] < a_train[i].quantile(0.1), a_train[i].quantile(0.1), a_train[i])
        a_train[i] = np.where(a_train[i] > a_train[i].quantile(0.9), a_train[i].quantile(0.9), a_train[i])
# Look again for skewness
for i, j in a_valid.skew().items():
    if j > 1.8:
        a_valid[i] = np.where(a_valid[i] < a_train[i].quantile(0.1), a_train[i].quantile(0.1), a_valid[i])
        a_valid[i] = np.where(a_valid[i] > a_train[i].quantile(0.9), a_train[i].quantile(0.9), a_valid[i])
a_train.skew()
a_train.skew()
```

```python
# handle date time 


a_train['Year'] = a_train['Date'].dt.year
a_train['Month'] = a_train['Date'].dt.month
a_train['Day'] = a_train['Date'].dt.day

a_valid['Year'] = a_valid['Date'].dt.year
a_valid['Month'] = a_valid['Date'].dt.month
a_valid['Day'] = a_valid['Date'].dt.day

a_train.corr()['Price'].sort_values()
```
